# Introduction to Pandas

# 2. Creando la serie Pandas

## **Serie Pandas**

Una serie de Pandas es un objeto similar a una matriz *unidimensional* que puede contener muchos tipos de datos, como números o cadenas, y tiene una opción para proporcionar etiquetas de eje.

### **Diferencia entre NumPy ndarrays y Pandas Series**

1. Una de las principales diferencias entre la serie Pandas y NumPy ndarrays es que puede asignar una etiqueta de índice a cada elemento de la serie Pandas. En otras palabras, puedes nombrar los índices de tu Pandas Series como quieras.
2. Otra gran diferencia entre Pandas Series y NumPy ndarrays es que Pandas Series puede contener datos de diferentes tipos.

```python
#importing pandas
import pandas as pd
```

```python
groceries = pd.Series(data = [20,6,'Yes', 'No'], index = ['eggs', 'apples', 'milk', 'bread'])
groceries
```

Vemos que Pandas Series se muestra con los índices en la primera columna y los datos en la segunda columna. Observe que los datos no están indexados de 0 a 3, sino que están indexados con los nombres de los alimentos que ingresamos, es decir, huevos, manzanas, etc. Además, observe que los datos en nuestra serie Pandas tienen números enteros y cadenas.

## Atributos de impresión de una Serie

```python
print('Groceries tiene la forma:', groceries.shape)
print('Groceries tiene la dimensión:', groceries.ndim)
print('Groceries tiene un total de', groceries.size, 'elements')
```

```python
print('Los datos en Groceries son:', groceries.values)
print('Los indices en Groceries son:', groceries.index)
```

También puede verificar si el índice deseado está en la Serie o no, como un diccionario donde usamos claves en lugar de índices.

```python
print('Los datos en Groceries son:', groceries.values)
print('Verifica si eggs está en Groceries:','eggs' in groceries)
print('Verifica si bananas está en Groceries:','bananas' in groceries)
```

# 3. Acceder y eliminar elementos en la serie Pandas

La serie Pandas se puede considerar como una combinación de las listas de estructuras de datos incorporadas de Python y los diccionarios, ya que podemos acceder a los elementos de una serie Pandas accediendo a ellos con un índice basado en números enteros como lo hacemos con las listas que comienzan con el índice 0 o podemos acceder elemento por valor de índice en forma de etiqueta como una clave en los diccionarios.

## Accediendo a elementos

```python
# We access elements in Groceries using index labels:# We use a single index label
print('How many eggs do we need to buy:', groceries['eggs'])
print()

# we can access multiple index labels
print('Do we need milk and bread:\n', groceries[['milk', 'bread']])
print()

# we use loc to access multiple index labels
print('How many eggs and apples do we need to buy:\n', groceries.loc[['eggs', 'apples']])
print()

# We access elements in Groceries using numerical indices:# we use multiple numerical indices
print('How many eggs and apples do we need to buy:\n',  groceries[[0, 1]])
print()

# We use a negative numerical index
print('Do we need bread:\n', groceries[-1])
print()

# We use a single numerical index
print('How many eggs do we need to buy:', groceries[0])
print()
# we use iloc to access multiple numerical indices
print('Do we need milk and bread:\n', groceries.iloc[[2, 3]])
```

## Modificando un elemento

```python
# We display the original grocery list
print('Original Grocery List:\n', groceries)

# We change the number of eggs to 2
groceries['eggs'] = 2

# We display the changed grocery list
print()
print('Modified Grocery List:\n', groceries)
```

## Eliminar un elemento

```python
# We display the original grocery list
print('Original Grocery List:\n', groceries)

# We remove apples from our grocery list. The drop function removes elements out of place
print()
print('We remove apples (out of place):\n', groceries.drop('apples'))

# When we remove elements out of place the original Series remains intact. To see this# we display our grocery list again
print()
print('Grocery List after removing apples out of place:\n', groceries)
```

Podemos eliminar elementos de una serie Pandas configurando la palabra clave inplace en True en el método .drop() .

```python
# We display the original grocery list
print('Original Grocery List:\n', groceries)

# We remove apples from our grocery list in place by setting the inplace keyword to True
groceries.drop('apples', inplace =True)

# When we remove elements in place the original Series its modified. To see this# we display our grocery list again
print()
print('Grocery List after removing apples in place:\n', groceries)
```

# 3. Serie Operaciones aritméticas en Pandas

Al igual que con NumPy ndarrays, podemos realizar operaciones aritméticas por elementos en Pandas Series.

```python
fruits= pd.Series(data = [10, 6, 3,], index = ['apples', 'oranges', 'bananas'])
fruits
```

```python
# We print fruits for reference
print('Original grocery list of fruits:\n ', fruits)

# We perform basic element-wise operations using arithmetic symbols
print()
print('fruits + 2:\n', fruits + 2)# We add 2 to each item in fruits
print()
print('fruits - 2:\n', fruits - 2)# We subtract 2 to each item in fruits
print()
print('fruits * 2:\n', fruits * 2)# We multiply each item in fruits by 2
print()
print('fruits / 2:\n', fruits / 2)# We divide each item in fruits by 2
print()
```

**También podemos usar funciones aritméticas de NumPy en Series**

```python
# We import NumPy as np to be able to use the mathematical functions
import numpy as np

# We print fruits for reference
print('Original grocery list of fruits:\n', fruits)

# We apply different mathematical functions to all elements of fruits
print()
print('EXP(X) = \n', np.exp(fruits))
print()
print('SQRT(X) =\n', np.sqrt(fruits))
print()
print('POW(X,2) =\n',np.power(fruits,2))# We raise all elements of fruits to the power of 2
```

```python
# We can also perform this above operations on selected elements

# We print fruits for reference
print('Original grocery list of fruits:\n ', fruits)
print()

# We add 2 only to the bananas
print('Amount of bananas + 2 = ', fruits['bananas'] + 2)
print()

# We subtract 2 from apples
print('Amount of apples - 2 = ', fruits.iloc[0] - 2)
print()

# We multiply apples and oranges by 2
print('We double the amount of apples and oranges:\n', fruits[['apples', 'oranges']] * 2)
print()

# We divide apples and oranges by 2
print('We half the amount of apples and oranges:\n', fruits.loc[['apples', 'oranges']] / 2)
```

# 4. Creando marcos de datos de Pandas

Pandas DataFrames son estructuras de datos bidimensionales con filas y columnas etiquetadas, que pueden contener muchos tipos de datos. Si está familiarizado con Excel, puede pensar que Pandas DataFrames es similar a una hoja de cálculo. Podemos crear Pandas DataFrames manualmente o cargando datos desde un archivo.

## **Crear un DataFrame manualmente**

Comenzaremos creando un DataFrame manualmente a partir de un diccionario de Pandas Series. Es un proceso de dos pasos:

1. El primer paso es crear el diccionario de Pandas Series.
2. Una vez creado el diccionario, podemos pasarlo a la función `pd.DataFrame()` .

```python
# We import Pandas as pd into Python
import pandas as pd

# We create a dictionary of Pandas Series
items = {'Bob' : pd.Series(data = [245, 25, 55], index = ['bike', 'pants', 'watch']),
         'Alice' : pd.Series(data = [40, 110, 500, 45], index = ['book', 'glasses', 'bike', 'pants'])}

# We print the type of items to see that it is a dictionary
print(type(items))
items
```

```python
# We create a Pandas DataFrame by passing it a dictionary of Pandas Series
shopping_carts = pd.DataFrame(items)

# We display the DataFrame
shopping_carts
```

Hay varias cosas a tener en cuenta aquí, como se explica a continuación:

1. Vemos que los DataFrames se muestran en forma tabular, muy parecido a una hoja de cálculo de Excel, con las etiquetas de filas y columnas en **negrita**.
2. Además, observe que las etiquetas de fila del DataFrame se crean a partir de la unión de las etiquetas de índice de las dos series Pandas que utilizamos para construir el diccionario. Y las etiquetas de las columnas del DataFrame se toman de las *claves* del diccionario.
3. Otra cosa a tener en cuenta es que las columnas están ordenadas alfabéticamente y no en el orden indicado en el diccionario. Más adelante veremos que esto no sucederá cuando carguemos datos en un DataFrame desde un archivo de datos.
4. Lo último que queremos señalar es que vemos aparecer algunos valores `NaN` en el DataFrame. `NaN` significa *No es un número*, y es la forma de Pandas de indicar que no tiene un valor para ese índice de fila y columna en particular. Por ejemplo, si miramos la columna de Alice, vemos que tiene `NaN` en el índice de observación. Puedes ver por qué es así consultando el diccionario que creamos al principio. Vemos claramente que el diccionario no tiene ningún elemento para los relojes con la etiqueta Alice. Entonces, cada vez que se crea un DataFrame, si una columna en particular no tiene valores para un índice de fila en particular, Pandas pondrá un valor `NaN` allí.
5. Si introducimos estos datos en un algoritmo de aprendizaje automático, primero tendremos que eliminar estos valores `NaN` .

En el ejemplo anterior, creamos un Pandas DataFrame a partir de un diccionario de la serie Pandas que tenía índices claramente definidos. Si no proporcionamos etiquetas de índice a la serie Pandas, Pandas utilizará índices de filas numéricos cuando cree el DataFrame.

```python
# We create a dictionary of Pandas Series without indexes
data = {'Bob' : pd.Series([245, 25, 55]),
        'Alice' : pd.Series([40, 110, 500, 45])}

# We create a DataFrame
df = pd.DataFrame(data)

# We display the DataFrame
df
```

# 5. Acceder, modificar y eliminar elementos en pandas DataFrames

```python
items = [{'bikes' : 15,'pants': 20,'watches': 35},{'bikes' : 12,'pants':30,'watches':40,'glass':10}]
store_items = pd.DataFrame(items,index=['store 1','store 2'])
store_items
```

## Accediendo a elementos

```python
# We access rows, columns and elements using labels
print()
print('How many bikes are in each store:\n', store_items[['bikes']])
print()
print('How many bikes and pants are in each store:\n', store_items[['bikes', 'pants']])
print()
print('What items are in Store 1:\n', store_items.loc[['store 1']])
print()
print('How many bikes are in Store 2:', store_items['bikes']['store 2'])
```

## Agregar nueva columna

```python
# We add a new column named shirts to our store_items DataFrame indicating the number of# shirts in stock at each store. We will put 15 shirts in store 1 and 2 shirts in store 2
store_items['shirts'] = [15,2]

# We display the modified DataFrame
store_items
```

```python
# We make a new column called suits by adding the number of shirts and pants
store_items['suits'] = store_items['pants'] + store_items['shirts']

# We display the modified DataFrame
store_items
```

```python
# We add a new column using data from particular rows in the watches column
store_items['new watches'] = store_items['watches'][1:]

# We display the modified DataFrame
store_items
```

```python
# We insert a new column with label shoes right before the column with numerical index 4
store_items.insert(4, 'shoes', [8,3])

# we display the modified DataFrame
store_items
```

## Agregar nueva fila

```python
# We create a dictionary from a list of Python dictionaries that will contain the number of different items at the new store
new_items = [{'bikes': 20, 'pants': 30, 'watches': 35, 'glasses': 4}]

# We create new DataFrame with the new_items and provide and index labeled store 3
new_store = pd.DataFrame(new_items, index = ['store 3'])

# We display the items at the new store
new_store
```

```python
# We append store 3 to our store_items DataFrame
store_items = store_items.append(new_store)

# We display the modified DataFrame
store_items
```

## Eliminar filas y columnas de un marco de datos

```python
# We remove the new watches column
store_items.pop('new watches')

# we display the modified DataFrame
store_items
```

```python
# We remove the watches and shoes columns
store_items = store_items.drop(['watches', 'shoes'], axis = 1)

# we display the modified DataFrame
store_items
```

```python
# We remove the store 2 and store 1 rows
store_items = store_items.drop(['store 2', 'store 1'], axis = 0)

# we display the modified DataFrame
store_items
```

## Modificar marco de datos

```python
# We change the column label bikes to hats
store_items = store_items.rename(columns = {'bikes': 'hats'})

# we display the modified DataFrame
store_items
```

```python
# We change the row label from store 3 to last store
store_items = store_items.rename(index = {'store 3': 'last store'})

# we display the modified DataFrame
store_items
```

```python
# We change the row index to be the data in the pants column
store_items = store_items.set_index('pants')

# we display the modified DataFrame
store_items
```

# 6. Lidiar con los valores perdidos

Antes de que podamos comenzar a entrenar nuestros algoritmos de aprendizaje con grandes conjuntos de datos, generalmente primero debemos limpiar los datos. Esto significa que necesitamos tener un método para detectar y corregir errores en nuestros datos. Si bien cualquier conjunto de datos determinado puede tener muchos tipos de datos incorrectos, como valores atípicos o valores incorrectos, el tipo de datos incorrectos que encontramos casi siempre son valores faltantes. Como vimos anteriormente, Pandas asigna valores NaN a los datos faltantes.

```python
import pandas as pd
# We create a list of Python dictionaries
items2 = [{'bikes': 20, 'pants': 30, 'watches': 35, 'shirts': 15, 'shoes':8, 'suits':45},
{'watches': 10, 'glasses': 50, 'bikes': 15, 'pants':5, 'shirts': 2, 'shoes':5, 'suits':7},
{'bikes': 20, 'pants': 30, 'watches': 35, 'glasses': 4, 'shoes':10}]

# We create a DataFrame  and provide the row index
store_items = pd.DataFrame(items2, index = ['store 1', 'store 2', 'store 3'])

# We display the DataFrame
store_items
```

## Contando el número de valores faltantes

```python
# We count the number of NaN values in store_items
x =  store_items.isnull().sum().sum()

# We print x
print('Number of NaN values in our DataFrame:', x)
```

Demostrando cómo funciona la línea de código anterior

```python
store_items.isnull()
```

```python
store_items.isnull().sum()
```

```python
store_items.isnull().sum().sum()
```

## Contando los valores totales que no son Nan

## Eliminando valores de NaN

Ahora que hemos aprendido cómo saber si nuestro conjunto de datos tiene valores NaN , el siguiente paso es decidir qué hacer con ellos. En general, tenemos dos opciones: podemos eliminar o reemplazar los valores NaN.

```python
# We drop any rows with NaN values
store_items.dropna(axis = 0)
```

```python
# We drop any columns with NaN values
store_items.dropna(axis = 1)
```

## Sustituyendo valores de NaN

```python
print("Original dataframe for refrence:")
store_items
```

```python
# We replace all NaN values with 0
store_items.fillna(0)
```

```python
# We replace NaN values with the previous value in the column
store_items.fillna(method = 'ffill', axis = 0)
```

```python
# We replace NaN values with the previous value in the row
store_items.fillna(method = 'ffill', axis = 1)
```

```python
# We replace NaN values with the next value in the column
store_items.fillna(method = 'backfill', axis = 0)
```

```python
# We replace NaN values with the next value in the row
store_items.fillna(method = 'backfill', axis = 1)
```

```python
# We replace NaN values by using linear interpolation(Estimating) using column values
store_items.interpolate(method = 'linear', axis = 0)
```

```python
# We replace NaN values by using linear interpolation(Estimating) using row values
store_items.interpolate(method = 'linear', axis = 1)
```
